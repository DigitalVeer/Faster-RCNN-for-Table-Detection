```python
import os
import pandas as pd
import re
import sys
import cv2


imageDir = "data/images/"
annotationFile = "data/train.csv"
outputFile = "train_annotations.txt"


data_train = pd.read_csv(annotationFile, names = ["filename", "xmin", "ymin", "xmax", "ymax", "class"])



```

```python
# for training
with open(outputFile, "w+") as f:
    for idx, row in data_train.iterrows():
        x1 = int(row['xmin'])
        x2 = int(row['xmax'])
        y1 = int(row['ymin'])
        y2 = int(row['ymax'])
        fileName = os.path.join(imageDir, row['filename'])
        className = row['class']
        f.write(fileName + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + className + '\n')


```

```python
testAannotationFile = "data/val.csv"
outputFile = "test_annotations.txt"

data_test = pd.read_csv(testAannotationFile, names = ["filename", "xmin", "ymin", "xmax", "ymax", "class"])
```
